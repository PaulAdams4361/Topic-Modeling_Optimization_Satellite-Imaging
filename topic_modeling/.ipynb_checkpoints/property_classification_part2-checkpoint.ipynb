Using the same dataset we would like you to populate the Property_Type column with one of the following values: Farm, Timberland, Ranch, Recreational, or Not Discernable, for properties greater than 180 acres. The determination will be from evaluation of key terms in the description. The chances are high that a description may contain terms from multiple classification types. Use proximity and frequency to assign a classification. This project may be more difficult, so feel free to ask questions and take liberties.


Key 

Farm: Farm, Crops, Row Crop, Pivot, Farming, Farmland, Bushels(IF Tillable Acres > 80% Total Acres)

Ranch: Ranch, Cattle, Grazing, Grassland, Pasture, 

Timberland: Timber, Forest, Woods, Wooded

Recreational: Hunting, Deer, Turkey, Recreational, Sportsman, Outdoorsman, Retreat

# Important Notes & Summary
Important to note is the conversion of "bushels" to "Not Discernable" for proerties that have tillable acres less than 80% of total acres. For this assignment, I did not make any imputations for missing data points in the tillable acres feature. This is something I would consult with the team to determine how best to handle these missing data points before imputing them on my own, since the results could be misleading. Therefore, since 96% of the tillable acres feature is missing, no "bushels" were converted to "Farm." As a result, there were originally no "Farm" values. After realizing this, I removed "bushels" from the possible categories and re-ran the data through the pipeline using a custom function. Under the assumption the clustering was reasonable and the term frequency-inverse document frequency was not over-fit (I limited the entire corpus of unique words available for topic analysis to 5% of all words (5% was approximately 487 words)), other Farm-related attributes surfaced. I think 487 words was probably excessive for topic modeling, but given the missingness of tillable acres, I felt it was necessary, especially since Farms is a minority class in the dataset, but one that is needed for business opportunity. The classification for farms increased from 19 to 198 after excluding "bushels" and using 487 words for the topic modeling.

```python
# creating a function to preprocess data. Kernel keeps crashing at umap because dataset it so large. splitting into quarters to process one-quarter at a time
def prep_data(input_data):
    X_1 = input_data.astype(str).tolist()
    X_2 = [' '.join(t.split()[0:150]) for t in X_1]
    X = np.array(X_2, dtype=object)[:, np.newaxis]

    learned_embeddings = []

    for line in X:
        learned_embeddings.append(embedder(line))

    # (1000,512)
    for i in range(0,len(learned_embeddings)):
        learned_embeddings[i] = tf.reshape(learned_embeddings[i], [512])

    embeddings = np.array(learned_embeddings)
    
    return embeddings
```

```python
embedding1 = prep_data(text1)
embedding2 = prep_data(text2)
embedding3 = prep_data(text3)
embedding4 = prep_data(text4)
```

```python
umap_embeddings1 = umap.UMAP(n_neighbors=15, n_components=5, metric='cosine').fit_transform(embedding1) # 512 DIM
umap_embeddings2 = umap.UMAP(n_neighbors=15, n_components=5, metric='cosine').fit_transform(embedding2) # 512 DIM
umap_embeddings3 = umap.UMAP(n_neighbors=15, n_components=5, metric='cosine').fit_transform(embedding3) # 512 DIM
umap_embeddings4 = umap.UMAP(n_neighbors=15, n_components=5, metric='cosine').fit_transform(embedding4) # 512 DIM
```

```python
def extract_top_n_words_per_topic(tf_idf, count, docs_per_topic, n=20):
    words = count.get_feature_names()
    labels = list(docs_per_topic.topic)
    tf_idf_transposed = tf_idf.T
    indices = tf_idf_transposed.argsort()[:, -n:]
    top_n_words = {label: [(words[j], tf_idf_transposed[i][j]) for j in indices[i]][::-1] for i, label in enumerate(labels)}
    return top_n_words

def extract_topic_sizes(df):
    topic_sizes = (df.groupby(['topic'])
                     .text
                     .count()
                     .reset_index()
                     .rename({"topic": "topic", "text": "Size"}, axis='columns')
                     .sort_values("Size", ascending=False))
    return topic_sizes

top_n_words = extract_top_n_words_per_topic(tf_idf, count, texts_per_topic, n=487)
topic_sizes = extract_topic_sizes(text_df)
topic_sizes
```

```python
def recode_categories(dataframe, bushels=False) -> pd.DataFrame:
    '''
    REQUIRES:
    original_data (farms.csv), top_n_words (top_n_words.pickle), topic_sizes (descriptions_with_labels.csv), text_df(), and dataframe (df2).
    By default, this function excludes "bushels" from the list of potential categories since so many tillable acres values are missing from farms.csv.
    The required pickles and csv files other than farms.csv are created in this notebook.
    '''
    
    if bushels==False:
        classifications = ['farm', 'crops', 'row crop', 'pivot', 'farming', 'farmland', 'ranch', 'cattle', 'grazing', 'grassland', 'pasture'
                  , 'timber', 'forest', 'woods', 'wooded', 'hunting', 'deer', 'turkey', 'recreational', 'sportsman', 'outdoorsman', 'retreat']
    if bushels==True:
        classifications = ['farm', 'crops', 'row crop', 'pivot', 'farming', 'farmland', 'bushels', 'ranch', 'cattle', 'grazing', 'grassland', 'pasture'
                  , 'timber', 'forest', 'woods', 'wooded', 'hunting', 'deer', 'turkey', 'recreational', 'sportsman', 'outdoorsman', 'retreat']


    discernable = pd.DataFrame()
    non_discernable = pd.DataFrame()
    a_list = []
    cluster = []
    
    for col in dataframe:
        first_index = dataframe[dataframe.loc[:,col].isin(classifications)].first_valid_index()
        if first_index is not None:
            cluster.append(col)
            a_list.append(dataframe.loc[first_index,col])

        else:
            cluster.append(col)
            a_list.append('not discernable')

    final_classification = pd.DataFrame({'topic':cluster, 'property_type':a_list})
    
    final_classification.iloc[0,1] = 'noise_cluster'

    text_df.to_csv('./text_df_final.csv')
    original_data.to_csv('./original_data.csv')

    text_df_final = text_df.merge(final_classification, on='topic')

    df_classified = pd.concat([text_df_final,original_data], axis=1).drop(['text','topic','text_id'], axis=1)
    
    farm = ['farm','crops','row crop','pivot','farming','farmland']
    #bushels = 'bushels'
    ranch = ['ranch','cattle','grazing','grassland','pasture']
    timberland = ['timber','forest','woods','wooded']
    recreational = ['hunting','deer','turkey','recreational','sportsman','outdoorsman','retreat']
    
    if bushels==True:
        for i in range(0,len(df_classified)):
            if df_classified.iloc[i,0].lower().strip() in farm:
                df_classified.iloc[i,0] = 'farm'

            elif (df_classified.iloc[i,0].lower().strip() == 'bushels') & (df_classified.loc[i,'tillable_acres'] > (0.8 * df_classified.loc[i,'acres'])):
                df_classified.iloc[i,0] = 'farm'

            elif df_classified.iloc[i,0].lower().strip() in ranch:
                df_classified.iloc[i,0] = 'ranch'

            elif df_classified.iloc[i,0].lower().strip() in timberland:
                df_classified.iloc[i,0] = 'timberland'

            elif df_classified.iloc[i,0].lower().strip() in recreational:
                df_classified.iloc[i,0] = 'recreational'

            else:
                df_classified.iloc[i,0] = 'not discernable'

        print('bushels == True')
        print('Percent of missing tillable acres: {}%'.format(round(df_classified['tillable_acres'].isna().sum()/len(df_classified)*100,2)))

        df_classified_bushels = df_classified

        return df_classified_bushels
        
    if bushels==False:
        for i in range(0,len(df_classified)):
            if df_classified.iloc[i,0].lower().strip() in farm:
                df_classified.iloc[i,0] = 'farm'

            elif df_classified.iloc[i,0].lower().strip() in ranch:
                df_classified.iloc[i,0] = 'ranch'

            elif df_classified.iloc[i,0].lower().strip() in timberland:
                df_classified.iloc[i,0] = 'timberland'

            elif df_classified.iloc[i,0].lower().strip() in recreational:
                df_classified.iloc[i,0] = 'recreational'

            else:
                df_classified.iloc[i,0] = 'not discernable'

        print('bushels == False')
        print('Percent of missing tillable acres: {}%'.format(round(df_classified['tillable_acres'].isna().sum()/len(df_classified)*100,2)))
            
        df_classified_no_bushels = df_classified
            
        return df_classified_no_bushels
```
